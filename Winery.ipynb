```python
# Creating a function which returns the uniquely labelled data
def label(df,heading):
    # replacing na values in college with No college 
    df[heading].fillna("Nothing present", inplace = True)

    # label_encoder object knows how to understand word labels. 
    label_encoder = preprocessing.LabelEncoder() 

    # Encode labels in column specified. 
    df[heading]= label_encoder.fit_transform(df[heading]) 

    #df['country'].unique()
    return df
```

```python
def preprocess(inpu):
    df = pd.read_csv(inpu)
    if inpu=='train.csv':
        a=label(df,'variety')
    else:
        a=df
    b=a
    var=b
    del a['review_title']
    del a['review_description']
    del var['user_name']
    del var['country']
    del var['designation']
    del var['points']
    del var['price']
    del var['province']
    del var['region_1']
    del var['region_2']
    del var['winery']
    de=pd.read_csv(inpu)
    v=label(de,'country')
    v=label(v,'winery')
    v=label(v,'region_1')
    v=label(v,'region_2')
    v=label(v,'province')
    v=label(v,'designation')
    v=label(v,'user_name')

    #replacing all NaN values with avg of that coulmn
    q=pd.read_csv('train.csv')
    q['price'].fillna(0, inplace = True)
    q['points'].fillna(0, inplace = True)
    av1=round(sum(np.array(q['price']))/len(q['price']))
    av2=round(sum(np.array(q['points']))/len(q['points']))
    v['price'].fillna(av1, inplace = True)
    v['points'].fillna(av2, inplace = True)
    del v['review_title']
    del v['review_description']
    if inpu=='train.csv':
        del v['variety']
    out=v

    #converting all values to float for easy calculation
    out = out.astype(float)
    var = var.astype(float)
    return(out,var)
```

```python
#defining the vaiable required to preform gnb
data_train,target_train=preprocess('train.csv')
data_test,target_test=preprocess('test.csv')
target_train=np.ravel(np.array(target_train))
data_train=np.array(data_train)
```

```python
"""this cell prints the prediction and the accuracy i am using the train data only for the testing, however in the next cell
i am using the same code to calculate the test result but cannot calculate the accuracy as the traget test data was 
not given"""
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
#create an object of the type GaussianNB
gnb = GaussianNB()
#train the algorithm on training data and predict using the testing data
pred = gnb.fit(data_train, target_train).predict(data_train)
#print(pred.tolist())
#print the accuracy score of the model
print(target_train)
print(pred)
print("Naive-Bayes accuracy : ",accuracy_score(target_train, pred, normalize = True))
```
